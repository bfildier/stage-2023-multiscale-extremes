```python
#Processing bunch of data
ds2_shape = (1984, 128, 48)
ds2_Prec = ds2["Prec"].stack(flat=('x', 'y', 'time')).to_numpy()
dist_SAM_Prec = cs.Distribution(name="SAM Precipitation", bintype = "invlogQ", nd = 6, fill_last_decade=True)
dist_SAM_Prec.computeDistribution(sample = ds2_Prec)
dist_SAM_Prec.storeSamplePoints(sample = ds2_Prec)

ds1_shape = (4, 62)
ds1_Prec = ds1["DAILY_precip"] #Precipitation daily average of the pixels with values equal or larger than 0 mm/h for each 1deg. x 1deg. grid points in one day
ds1_CondPrec= ds1["DAILY_conditionalprecip"][0,:,:]
ds1_FracPrec= ds1["DAILYprecipfrac"][0,:,:]
```

```python
## Prepare data

SAM_ranks = dist_SAM_Prec.ranks[:]
SAM_bin_locations = dist_SAM_Prec.bin_locations[:]
ranks=[]
loc = []
for r, bin_loc in zip(SAM_ranks, SAM_bin_locations) : 
    index_3d_x, index_3d_y = np.unravel_index(bin_loc, ds2_shape)[0], np.unravel_index(bin_loc, ds2_shape)[1]
    for x,y in zip(index_3d_x, index_3d_y):
        ranks.append(r)
        loc.append((x,y))
```

```python
filtered_coords = []
filtered_ranks = []

for coord,rank in zip(loc, ranks):
    x, y = coord
    if 30*32 <= x <= 32*32 and 2*32 <= y <= 4*32:
        filtered_coords.append(coord)
        filtered_ranks.append(rank)
```

```python
import numpy as np

# Define the bounds for each square
bounds_x = [30*32, 31*32, 32*32]
bounds_y = [2*32, 3*32,  4*32]

# Create empty lists to hold the maximum values and their corresponding coordinates
max_values = []
filtered_coord_max_values = []

# Iterate over each square and find the maximum value and its corresponding coordinate within it
for i in range(len(bounds_x)-1):
    for j in range(len(bounds_y)-1):
        # Filter the coordinates and their corresponding values based on the current square
        filtered_coords_square = [coord for coord in filtered_coords if (bounds_x[i] <= coord[0] <= bounds_x[i+1]) and (bounds_y[j] <= coord[1] <= bounds_y[j+1])]
        filtered_values_square = [filtered_ranks[filtered_coords.index(coord)] for coord in filtered_coords_square]
        
        # Find the maximum value within filtered_values_square
        max_value = np.max(filtered_values_square)
        
        # Find the coordinate with the maximum value within the square
        max_index = filtered_values_square.index(max_value)
        max_coord = filtered_coords_square[max_index]
        
        # Add the maximum value and its corresponding coordinate to the lists
        max_values.append(max_value)
        filtered_coord_max_values.append(max_coord)
```

```python
%matplotlib inline
from myFuncs import Normalize_IL as norm

norm = colors.FuncNorm((norm(nd=4)._transform_IL, norm(nd=4)._inverse_IL), vmin = 0)
ticks = [0, 90, 99, 99.9, 99.99]
ticklabels = ['{:.4f}'.format(t).rstrip('0').rstrip('.') for t in ticks]

fig, ax = plt.subplots(figsize=(6.5,4))
sc = ax.scatter([c[0] for c in filtered_coord_max_values], [c[1] for c in filtered_coord_max_values], c=max_values, cmap="Blues", norm=norm, s=30, alpha=0.8)
#im = ax.imshow(ds1_Prec[0,2:4,30:32], extent=[0, 3*32, 0, 2*32], aspect='auto', cmap='Blues', origin = 'lower')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Max Daily Precip')
cbar = fig.colorbar(sc)
ax.set_xlim(30*32,32*32)
cbar.set_label('Ranks')
cbar.set_ticks(ticks)
cbar.set_ticklabels(ticklabels)
plt.savefig("../figures/SAM_extremes_plotted_over_CACATOES_Daily_Precip.png", dpi= 600)
plt.tight_layout()
plt.show()
```
